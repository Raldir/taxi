```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('out/taxi_distributional_semantics_results.csv', skiprows=[1, 2, 3])
```

```python
df.head()
```

```python
df = df.loc[(df['buffer'] == 5) & (df['mode'] == 'reattach')]
```

```python
df = df.loc[(df['buffer'] == 5) & (df['mode'] == 'only_removal')]
```

```python
df_both = df.loc[(df['parent'] == 'poincare') & (df['family'] == 'fasttext')]
df_both
```

```python
df_parent = df.loc[(df['parent'] == 'poincare') & (df['family'] == '-')]
df_parent
```

```python
df_family = df.loc[(df['parent'] == '-') & (df['family'] == 'fasttext')]
df_family
```

```python
# Buffer: 5 | Mode: Reattach

plt.plot(df_both['iterations'], df_both['f1'], label='Both')
plt.plot(df_parent['iterations'], df_parent['f1'], alpha=0.7, label='Only Parent')
plt.plot(df_family['iterations'], df_family['f1'], alpha=0.7, label='Only Family')
plt.legend(loc='lower right')
plt.xlabel('Number of iterations')
plt.ylabel('F1 Score')
plt.title('Buffer: 5 | Mode: Reattach')
plt.axis([0, 10, 0.364, 0.369])
plt.show()
```

```python
# Buffer: 5 | Mode: Only Removal

plt.plot(df_both['iterations'], df_both['f1'], label='Both')
plt.plot(df_parent['iterations'], df_parent['f1'], alpha=0.7, label='Only Parent')
plt.plot(df_family['iterations'], df_family['f1'], alpha=0.7, label='Only Family')
plt.legend(loc='lower right')
plt.xlabel('Number of iterations')
plt.ylabel('F1 Score')
plt.title('Buffer: 5 | Mode: only_removal')
plt.axis([0, 10, 0.364, 0.369])
plt.show()
```
